```python
# Import library yang diperlukan
import pandas as pd
```

```python
# Nama file CSV
CSV_FILE = "motor_data.csv"

try:
    # Membaca file CSV menggunakan pandas
    data = pd.read_csv(CSV_FILE)
    
    # Menampilkan data dalam tabel
    print("Data from motor_data.csv:")
    display(data)  # Gunakan display() untuk tampilan Jupyter yang lebih baik
except FileNotFoundError:
    print(f"Error: File {CSV_FILE} not found.")
except Exception as e:
    print(f"Error reading CSV file: {e}")
```

```python
# Menyaring data hanya dengan StatusValue > 1800
filtered_data = data[data['StatusValue'] > 1500]

# Menampilkan data yang sudah difilter
print("Filtered Data (StatusValue > 1800):")
display(filtered_data)  # Gunakan display() untuk tampilan Jupyter yang lebih baik
```

```python
# Memeriksa nilai ModeValue yang bukan 0 dan ForwardValue yang bukan 2000
invalid_mode_forward = filtered_data[(filtered_data['ModeValue'] != 0) & (filtered_data['ForwardValue'] != 2000) & (filtered_data['SafetyValue'] != 2000)]

# Menampilkan hasil pemeriksaan ModeValue dan ForwardValue
if invalid_mode_forward.empty:
    print("Tidak ditemukan nilai ModeValue =! 0 dan ForwardValue =! 2000 dalam data dengan StatusValue =!2000.")
else:
    print("Ditemukan nilai ModeValue =! 0 dan ForwardValue =! 2000 dalam data dengan StatusValue =!2000:")
    display(invalid_mode_forward)
```

```python
# Menghapus kolom ModeValue, ForwardValue, dan SafetyValue dari filtered_data
cleaned_data = filtered_data.drop(columns=['ModeValue', 'ForwardValue', 'SafetyValue'])

# Menampilkan data yang telah dibersihkan
print("Data setelah menghapus kolom ModeValue, ForwardValue, dan SafetyValue:")
display(cleaned_data)
```

```python
# Memeriksa nilai unik pada ThrottleValue
unique_values = cleaned_data['ThrottleValue'].unique()

# Menampilkan semua nilai unik
print("Nilai unik pada ThrottleValue:")
for value in unique_values:
    print(value)
```

```python
# Memeriksa nilai unik pada TurnValue
unique_turn_values = cleaned_data['TurnValue'].unique()

# Menampilkan semua nilai unik
print("Nilai unik pada TurnValue:")
for value in unique_turn_values:
    print(value)
```

```python
# Menyaring data untuk nilai ThrottleValue >= 1000
filtered_throttle_data = cleaned_data[cleaned_data['ThrottleValue'] >= 1000]

# Menampilkan data hanya untuk baris dengan nilai unik ThrottleValue
unique_throttle_data = filtered_throttle_data[filtered_throttle_data['ThrottleValue'].isin(filtered_throttle_data['ThrottleValue'].unique())]

# Menampilkan data hasil filter
print("Data berdasarkan nilai unik pada ThrottleValue (ThrottleValue >= 1000):")
display(unique_throttle_data)
```

```python
# Menghapus kolom Timestamp dan StatusValue
final_data = unique_throttle_data.drop(columns=['Timestamp', 'StatusValue'])

# Nama file CSV baru
output_csv_file = "filtered_turn_throttle_data.csv"

# Menyimpan data hasil filter ke CSV
final_data.to_csv(output_csv_file, index=False)

print(f"Data hasil filter (hanya TurnValue dan ThrottleValue) telah disimpan ke file: {output_csv_file}")
```

```python
# Menghapus duplikasi berdasarkan pasangan TurnValue dan ThrottleValue
unique_pairs = final_data.drop_duplicates()

# Menampilkan data pasangan unik
print("Pasangan unik antara ThrottleValue dan TurnValue:")
display(unique_pairs)
```

```python
# Fungsi untuk mapping nilai seperti pada Arduino
def map_value(value, in_min, in_max, out_min, out_max):
    return (value - in_min) * (out_max - out_min) / (in_max - in_min) + out_min

# Data unik untuk ThrottleValue dan TurnValue
throttle_values = [1482, 1237, 1897]
turn_values = [1512, 1505, 1519]

# Mapping nilai ThrottleValue
throttle_map_values = [map_value(value, 1000, 2000, 0, 200) for value in throttle_values]

# Mapping nilai TurnValue
turn_map_values = [map_value(value, 1000, 2000, -100, 100) for value in turn_values]

# Menampilkan hasil mapping
print("ThrottleValue -> ThrottleMapValue:")
for original, mapped in zip(throttle_values, throttle_map_values):
    print(f"{original} -> {mapped:.2f}")

print("\nTurnValue -> TurnMapValue:")
for original, mapped in zip(turn_values, turn_map_values):
    print(f"{original} -> {mapped:.2f}")
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression

# Data dari pengujian
throttle_map_values = [96.40, 47.40, 179.40]
turn_map_values = [2.40, 1.00, 3.80]

# Mengonversi data menjadi array numpy
X = np.array(throttle_map_values).reshape(-1, 1)  # ThrottleMapValue sebagai variabel independen
y = np.array(turn_map_values)  # TurnMapValue sebagai variabel dependen

# Membuat model regresi linear
model = LinearRegression()
model.fit(X, y)

# Mendapatkan koefisien regresi
slope = model.coef_[0]  # Gradien (m)
intercept = model.intercept_  # Intersep (c)

# Menampilkan persamaan
print("Persamaan linear untuk TurnValueMap:")
print(f"TurnValueMap = {slope:.4f} * ThrottleMapValue + {intercept:.4f}")
```
